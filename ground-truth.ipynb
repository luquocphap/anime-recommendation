```python
rating = pd.read_csv("Dataset/rating.csv")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# rating = pd.read_csv("your_dataset.csv")

positive_rating = rating[(rating['rating'] == -1) | (rating['rating'] >= 5)]

train_list = []
test_list = []

for user_id, user_data in rating.groupby('user_id'):
    pos = positive_rating[positive_rating['user_id'] == user_id]

    # Không có tương tác dương → cho hết vào train
    if len(pos) == 0:
        train_list.append(user_data)
        continue

    # Chỉ có 1 tương tác dương → cho vào train (để tránh lỗi)
    if len(pos) == 1:
        pos_train = pos
        pos_test = pd.DataFrame(columns=user_data.columns)
    else:
        pos_train, pos_test = train_test_split(pos, test_size=0.2, random_state=42)

    remaining = user_data.drop(pos.index)
    train = pd.concat([remaining, pos_train])

    train_list.append(train)
    if len(pos_test) > 0:
        test_list.append(pos_test)

train_rating = pd.concat(train_list).reset_index(drop=True)
test_rating = pd.concat(test_list).reset_index(drop=True)

print("Train size:", len(train_rating))
print("Test size:", len(test_rating))
print(f"Số user trong test: {test_rating['user_id'].nunique()}")
```
